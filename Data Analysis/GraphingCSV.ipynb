```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file
# Replace 'data.csv' with the path to your actual CSV file
df = pd.read_csv("C:\Code\\310\\experiment2.csv")

# Convert time from ms to seconds for better readability (optional)
df['time_s'] = df['time_ms'] / 1000.0

# Plot 1: Position (steps) vs Time
plt.figure(figsize=(10, 5))
plt.plot(df['time_s'], df['position_step'], label='Position (steps)', color='b')
plt.xlabel("Time (s)")
plt.ylabel("Position (steps)")
plt.title("Stepper Position Over Time")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Plot 2: Rotary sensor position (deg) vs Time
plt.figure(figsize=(10, 5))
plt.plot(df['time_s'], df['position_deg'], label='Rotary Sensor (deg)', color='r')
plt.xlabel("Time (s)")
plt.ylabel("Position (deg)")
plt.title("Rotary Sensor Position Over Time")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
